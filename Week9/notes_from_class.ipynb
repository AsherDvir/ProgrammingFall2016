```r
f <- function(x, param){
    return(param[1]*(x - param[2])^3)
}

f_prime <- function(x, param){
    return(param[1]*3*(x - param[2])^2)
}
```

```r
newton <- function(f, f_prime, x0, param, tol=1e-1, maxits=10){
    x1 <- 0
    norm <- 1e9
    
    i <- 1
    
    while(i <= maxits & norm > tol){
        x1 <- x0 - f(x0, param)/f_prime(x0, param)
        norm <- abs(x1 - x0)
        x0 <- x1
        i <- i + 1
        
        #if(norm <= tol){
        #    break
        #}
        if(i == maxits){
            print("maxits")
        }
    }
    return(x1)
}
```

```r
newton(f, f_prime, 2.0, c(1.0, 0.0), tol=1e-3, maxits=50)
```

```r
F <- function(x, param){
    return(param[1]/4*(x - param[2])^4)
}
optim(2.0, F, param=c(1.0, 0.0))
```

```r
system.time(optim(100.0, F, param=c(1.0, 0.0)))
system.time(newton(f, f_prime, 100.0, c(1.0, 0.0), tol=1e-3, maxits=50))
```
