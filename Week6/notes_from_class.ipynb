```r
n <- 4

A <- matrix(0, nrow=n, ncol=n)

for(i in 1:n){
    for(j in 1:n){
        if(i != j & (i == 1 | j == 1)){
            A[i, j] <- 1
        }
    }
}
plot(graph_from_adjacency_matrix(A))
```

```r
n <- 4
p <- 1.0
A <- matrix(0, nrow=n, ncol=n)

for(i in 1:n){
    for(j in 1:i){
        if(i != j){
            if(runif(1) <= p){
                A[i, j] <- 1
            }
        }
    }
}
A <- A + t(A)
#plot(graph_from_adjacency_matrix(A))
```

```r
n <- 10000
p <- 0.5


A <- matrix(0, nrow=n, ncol=n)

tic()
A[lower.tri(A)] <- sample(c(0, 1), size=as.integer(n*(n-1)/2),
                          replace=T, prob=c(1-p, p))
toc()

A <- matrix(0, nrow=n, ncol=n)

tic()
for(i in 2:n){
    for(j in 1:(i-1)){
        if(runif(1) <= p){
            A[i, j] <- 1
        }
    }
}
toc()

A <- A + t(A)
A
#plot(graph_from_adjacency_matrix(A))
```
